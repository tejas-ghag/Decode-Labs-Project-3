```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
```

# TASK 1 : Understanding the Dataset

```python
print("First 5 Rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

# TASK 2 : Data Cleaning

```python
# Checking Missing Values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Checking Duplicate Values
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
# Removing Duplicate Rows
df = df.drop_duplicates()
print("Duplicate Rows Removed Successfully")
```

```python
# Removing Rows with Missing Reviews
df = df.dropna(subset=["Review Text"])
print("Dataset Cleaned Successfully")
```

# TASK 3 : EDA 

```python
print("Average Rating:")
print(df["Rating"].mean())

print("Average Age:")
print(df["Age"].mean())

print("Average Review Length:")
print(df["review_length"].mean())
```

```python
print("Reviews per Department:")
print(df["Department Name"].value_counts())
```

```python
print("Top Clothing Categories:")
print(df["Class Name"].value_counts().head())
```

```python
# Create Review Length Column
df["review_length"] = df["Review Text"].apply(len)
print(df.head())
```

```python
print("Recommendation Count:")
print(df["Recommended IND"].value_counts())
```

### EDA Insights

```python
print("1. Customer ratings are generally positive.")
print("2. Some departments receive more reviews than others.")
print("3. Review lengths vary among customers.")
print("4. Recommended products tend to have higher ratings.")
print("5. Customer feedback helps understand customer preferences.")
```

# TASK 4 : Data Visaulization

### Rating Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Rating", data=df)
plt.title("Rating Distribution")
plt.show()
```

### Recommendation Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Recommended IND", data=df)
plt.title("Recommendation Distribution")
plt.show()
```

### Customer Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20)
plt.title("Customer Age Distribution")
plt.show()
```

### Department-wise Ratings

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    x="Department Name",
    y="Rating",
    data=df
)
plt.title("Department-wise Ratings")
plt.xticks(rotation=45)
plt.show()
```

### WordCloud Visualization

```python
text = " ".join(df["Review Text"].astype(str))
wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color="white"
).generate(text)

plt.figure(figsize=(12,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Most Common Words in Reviews")
plt.show()
```

# TASK 5 :  Customer Review Analysis & Insights

### Recommended vs Not Recommended Products

```python
recommended = df[df["Recommended IND"] == 1]
not_recommended = df[df["Recommended IND"] == 0]

print("Recommended Reviews:" , len(recommended))
print("Not Recommended Reviews:",len(not_recommended))
```

### Average Rating of Recommended Products

```python
print("Average Rating of Recommended Products:",recommended["Rating"].mean())
```

### Most Reviewed Department

```python
print("Most Reviewed Department:")
print(df["Department Name"].value_counts().idxmax())
```

### Highest Rated Department

```python
print("Highest Rated Department:")
print(
    df.groupby("Department Name")["Rating"]
      .mean()
      .idxmax()
)
```
