```python
import pandas as pd
```

```python
df = pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
```

## TASK 1 : Understanding the Dataset

```python
print("First 5 Rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

## TASK 2 : Data Cleaning

```python
# Checking Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Checking Duplicate Values
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# Removing Duplicate Rows
df = df.drop_duplicates()
```

```python
# Removing Rows with Missing Reviews
df = df.dropna(subset=["Review Text"])
print("\nDataset Cleaned Successfully")
```

## TASK 3 : EDA 

### Mean of data

```python
print("\nAverage Rating:")
print(df["Rating"].mean())

print("\nAverage Age:")
print(df["Age"].mean())

print("\nAverage Review Length:")
print(df["review_length"].mean())
```

```python
print("\nReviews per Department:")
print(df["Department Name"].value_counts())
```

```python
print("\nTop Clothing Categories:")
print(df["Class Name"].value_counts().head())
```

```python
# Create Review Length Column
df["review_length"] = df["Review Text"].apply(len)
print(df.head())
```

```python
print("\nRecommendation Count:")
print(df["Recommended IND"].value_counts())
```
